```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,root_mean_squared_error
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/gym_members_exercise_tracking.csv")
df.head()
```

```python
df.info()
```

```python
df["Gender"] = df["Gender"].map({"Male": 1, "Female": 0})
```

```python
df
```

```python
df['Workout_Type'].unique()
df['Workout_Type']=df["Workout_Type"].map({"Yoga": 0, "HIIT": 1,"Cardio": 2,"Strength": 3})
```

```python
df
```

```python
df['Experience_Level'].unique()
```

```python
df.isnull().mean()
```

```python
inp=df.drop(columns=['Experience_Level'])
out=df['Experience_Level']
```

```python
xtr,xte,ytr,yte=train_test_split(inp,out,test_size=0.2,random_state=42)
```

```python
rf= RandomForestClassifier()
rf.fit(inp,out)
```

```python
df.columns
```

```python
samir_parameters={
    "n_estimators":[200,250,300],
    "criterion":["gini","entropy"],
    "max_depth":[5,6,7,8,9],
    "min_samples_split":[3,5,10],
    "min_samples_leaf":[1,2,4]
                 }

gs=GridSearchCV(rf,param_grid=samir_parameters,cv=5,verbose=2)
```

```python
gs.fit(xtr,ytr)
```

```python
print(gs.best_estimator_)
```

```python
print(gs.best_score_)
```

```python
model=gs.best_estimator_
```

```python
model.fit(xtr,ytr)
```

```python
pred=model.predict(xte)
accuracy_score=accuracy_score(yte,pred)
print(accuracy_score)
```
